```python
with  open(path, 'r') as f:
    lines = f.readlines()
    header = lines[0]
    dataset = {}
    header_arr = []
    contents = header.split('|')[1:-1]
    for content in contents:
        trimed_content = content.strip()
        header_arr.append(trimed_content)
        dataset[trimed_content] = []
    
    for line in lines[3:]:
        contents = line.split('|')[1:-1]
        if len(contents) == len(header_arr):
            i = 0
            for content in contents:
                trimed_content = content.strip()
                try:
                    header_trimed_content = header_arr[i]
                    dataset[header_trimed_content].append(trimed_content)
                except:
                    pass
                i+=1

    f.close()

    # for line in lines[:5]:
    #     print(line)
```

```python
def generating_codrinate():
    max_long = 45**10
    max_Llattude = 90**10
    Longitude = (random.randint(10000000000000, max_long)/max_long) *100 *random.randint(-2,1)
    latitude = (random.randint(10000000000000, max_Llattude)/max_Llattude) *100*random.randint(-2,1)
    if latitude and Longitude:
        return Longitude, latitude
    else:
        return generating_codrinate()
```

```python
desease_predicton_dictionary = {'disease':[], 'Longitude':[], 'latitude':[], 'occupaton':[], 'gender':[], 'Age':[], 'symptoms':[]}
```

```python
for index, Presentation in enumerate(combined_data.Disease):
    disease = combined_data.Category[index]
    occupaton =  combined_data['Occupation Most Affected'][index]
    age = combined_data['Age Range'][index].split(' ')
    all_gender = ['Male', 'Female']
    
    selected_gender = combined_data.Gender[index]
    text = combined_data['Clinical Presentation'][index]

    symptom_modification = text.split(',')
    coppied_symptom = []
    for item in symptom_modification:
        symptom_selected = random.choice(symptom_modification)
        coppied_symptom.append(symptom_selected)
        symptom_modification.remove(symptom_selected)

    if len(coppied_symptom) >6:
        coppied_symptom = coppied_symptom[:6]

    ds_symptom_arr =','.join(coppied_symptom)
    
    lowerAge, upperAge = age[0].split('-')
    lowerAge, upperAge = int(lowerAge), int(upperAge)
    medium_age = (lowerAge + upperAge)/2
    for i in range(0, int(medium_age)):
        age = random.randint(lowerAge-1, upperAge)
        Longitude, latitude = generating_codrinate()
        gender = ''
        if selected_gender == 'Any':
            gender = random.choice(all_gender)
        else:
            gender = selected_gender
        region = random.choice(regions_array)
        # occupaton = random.choice(occupaton_array)

        desease_predicton_dictionary['gender'].append(gender)
        desease_predicton_dictionary['occupaton'].append(occupaton)
        desease_predicton_dictionary['Age'].append(age)
        desease_predicton_dictionary['disease'].append(disease)
        desease_predicton_dictionary['Longitude'].append(Longitude)
        desease_predicton_dictionary['latitude'].append(latitude)
        desease_predicton_dictionary['symptoms'].append(ds_symptom_arr)
```

```python
symptoms_DICT = {}
for index, symptom in enumerate(list(set(symptoms))):
    symptoms_DICT[symptom.strip()] = index
```

```python
for model_index, symptom in enumerate(modeling_dataset.symptoms):
    symptoms_index = ''
    symptom_modification = symptom.split(',')
    for item in symptom_modification:
        symptom_value = symptoms_DICT[item.strip()]
        if symptom_value < 10:
            symptom_value= '00'+str(symptom_value)
        elif symptom_value < 100:
            symptom_value= '0'+str(symptom_value)
        else:
            symptom_value= str(symptom_value)
        symptoms_index += symptom_value + '00'
    modeling_dataset.symptoms[model_index] = int(symptoms_index)
```
